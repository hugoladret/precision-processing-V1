```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects

import MotionClouds as mc
import imageio


import sys
sys.path.append('../experimental')
import pipeline_params as prm
```

# Figure 1 : MotionClouds

```python
def generate_cloud(theta, b_theta, phase,
                   N_X, N_Y, seed, contrast=1.,
                   transition=False):
    
    
    fx, fy, ft = mc.get_grids(N_X, N_Y, 1)
    disk = mc.frequency_radius(fx, fy, ft) < .5

    if b_theta == 0 : 
        mc_i = mc.envelope_gabor(fx, fy, ft,
                                 V_X=0., V_Y=0., B_V=0.,
                                 sf_0=sf_0, B_sf=B_sf,
                                 theta=0, B_theta=b_theta)
        mc_i = np.rot90(mc_i)
    else :
        mc_i = mc.envelope_gabor(fx, fy, ft,
                                 V_X=0., V_Y=0., B_V=0.,
                                 sf_0=sf_0, B_sf=B_sf,
                                 theta=theta, B_theta=b_theta)
      
    im_ = np.zeros((N_X, N_Y, 1))
    im_ += mc.rectif(mc.random_cloud(mc_i, seed=seed),
                     contrast=2)
    im_ += -.5
    return im_[:,:,0]
```

```python
# Precisions, here we use 0.05 as an endpoint and not 0 because of a weird matplotlib behaviour
# todo : investigate, maybe it comes from np.rot90 ?
B_thetas = np.linspace(np.pi/2, 0.05, 8)/ 2.5

colors = prm.colors

# Spatial frequency, in cpd
sf = 0.0616279 # see stimulation notebooks
sf_0 = sf
B_sf = sf_0


fig, axs = plt.subplots(figsize = (16,8), nrows = 2, ncols = 4,
                        gridspec_kw = {'wspace':0.01, 'hspace':0.05})

theta = np.pi/4
N_X, N_Y = 512, 512

for ibt in range(0,4) :
    ax = axs[0, ibt]
    img = generate_cloud(theta = theta, b_theta = B_thetas[ibt], phase = 0,
                       N_X = N_X, N_Y = N_Y, seed = 42, contrast=1.,
                       transition=False)
    im = ax.imshow(img, cmap = 'gray', interpolation = 'bilinear')
    im.set_clim(-1,1)
    ax.set_yticks([])
    ax.set_xticks([])
    ax.set_aspect('equal')

    for edge in ['top', 'bottom', 'left', 'right'] :
        ax.spines[edge].set_color(colors[ibt])
        ax.spines[edge].set_linewidth(2)  

    
for ibt in range(0,4) :
    ax = axs[1, ibt]
    img = generate_cloud(theta = theta, b_theta = B_thetas[ibt+4], phase = 0,
                       N_X = N_X, N_Y = N_Y, seed = 42, contrast=1.,
                       transition=False)
    im = ax.imshow(img, cmap = 'gray', interpolation = 'bilinear')
    im.set_clim(-1,1)
    ax.set_yticks([])
    ax.set_xticks([])
    ax.set_aspect('equal')

    for edge in ['top', 'bottom', 'left', 'right'] :
        ax.spines[edge].set_color(colors[ibt+4])
        ax.spines[edge].set_linewidth(2)  


#fig.savefig('./output/fig_MC.pdf', bbox_inches = 'tight', transparent = True, dpi = 130)
```
