# Figure 1 : Naka Rushton curves

```python
import sys
sys.path.append('../experimental')
```

```python
import neuron_list as nlist
import pipeline_params as prm
```

```python

figsize = (5,5)

beg_psth = prm.beg_psth
end_psth = prm.end_psth
binsize = prm.binsize
B_thetas = prm.B_thetas
B_thetas = B_thetas * 180/np.pi

colors = plt.cm.inferno(np.linspace(.2, .9, len(B_thetas)))[::-1]
n_bin = (((end_psth) - (beg_psth)) * 1000) / binsize
```

```python
def nakarushton(x, rmax, c50, b, n):
    nkr = b + (rmax-b) * ((x**n) / (x**n + c50**n))
    return nkr


def norm_data(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

```python
def make_NKR_monoaxis(figsize, do_save, filename,
                      cv_fit, cv_raw,
                      hwhh_fit, hwhh_raw,
                      B_thetas):
    lw = 3
    s = 30
    fig, axs = plt.subplots(figsize=figsize, nrows = 2)

    # Rmax
    ax = axs[0]
    nkr = nakarushton(np.linspace(0, 1, 1000),
                      cv_fit['rmax'], cv_fit['c50'], cv_fit['b'], cv_fit['n'])
    ax.plot(np.linspace(np.min(B_thetas), np.max(B_thetas), 1000),
            nkr,
            c=r'#3C5488FF', linewidth=lw, linestyle=(0, (3, 1)), zorder=5)
    ax.scatter(B_thetas, cv_raw,
               facecolor=colors, edgecolor='k', marker='o', s=s, zorder=20)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks(np.round(np.linspace(np.min(B_thetas), np.max(B_thetas),
                           3, endpoint = True), 1 ))
    ax.set_xticklabels([])
    ax.text(x = .05 , y = .8 , s = 'CV', color = r'#3C5488FF', transform = ax.transAxes,
           fontsize = 15)
    #ax.set_ylabel('CV', fontsize=18)
    ax.set_yticks(np.round(np.linspace(np.min(cv_raw), np.max(cv_raw),
                           3, endpoint = True), 2 ))
    ax.set_ylim(np.min(cv_raw) - np.min(cv_raw) * .1,
                np.max(cv_raw) + np.max(cv_raw) * .1,)
    ax.tick_params(axis='both', which='major', labelsize=13)
    
    # HWHH
    ax = axs[1]
    nkr = nakarushton(np.linspace(0, 1, 1000),
                      hwhh_fit['rmax'], hwhh_fit['c50'], hwhh_fit['b'], hwhh_fit['n'])
    ax.plot(np.linspace(np.min(B_thetas), np.max(B_thetas), 1000),
            nkr,
            c=r'#00a087', linewidth=lw, linestyle=(0, (3, 1)), zorder=5)
    ax.scatter(B_thetas, hwhh_raw,
               facecolor=colors, edgecolor=r'k', marker='o', s=s, zorder=20)
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xlabel(r'B$_\theta (°)$', y=.85, fontsize=16)
    ax.text(x = .05 , y = .8 , s = 'HWHH(°)', color = r'#00a087', transform = ax.transAxes,
           fontsize = 15)
    #ax.set_ylabel('HWHH (°)', fontsize=18)
    ax.set_yticks(np.round(np.linspace(np.min(hwhh_raw), np.max(hwhh_raw),
                           3, endpoint = True), 2 ))
    ax.set_ylim(np.min(hwhh_raw) - np.min(hwhh_raw) * .1,
                np.max(hwhh_raw) + np.max(hwhh_raw) * .1,)
    
    ax.set_xticks(np.round(np.linspace(np.min(B_thetas), np.max(B_thetas),
                           3, endpoint = True), 2 ))
    ax.tick_params(axis='both', which='major', labelsize=13)

    
    
    if do_save:
        fig.savefig('./output/fig_1_NKR_%s.pdf' %  filename,
                    bbox_inches='tight', dpi=200, transparent=True)

    plt.show()
```

```python
grouping_path = '../experimental/results/testgroup21092020/clusters'
for cluster_path in nlist.lst:
    figsize = (2.5,2.5)
    folder_path = '_'.join(cluster_path.split('_')[:2])
    load_folder = '%s/%s' % (grouping_path, cluster_path)

    fitted_TC = np.load(load_folder + '/0.000_plot_neurometric_fitted_TC.npy')

    bthetas_fit = np.load(
        load_folder + '/0.000_plot_neurometric_Btheta_fits.npy')
    phi_fit = np.load(load_folder + '/0.000_phi_fit.npy', allow_pickle=True)
    phi_fit = phi_fit.item()
    phi_r2 = np.load(load_folder + '/0.000_phi_fitr2.npy')
    
    CVs = np.load(load_folder + '/0.000_cirvar.npy')
    fit_cv = np.load(load_folder + '/0.000_cirvar_fit.npy', allow_pickle = True)
    fit_cv = fit_cv.item()
    cv_r2 = np.load(load_folder + '/0.000_cirvar_fitr2.npy')

    print(cluster_path)
    print('HWHH n = %.2f ; B50 = %.2f°, r^2 = %.2f' %
          (phi_fit['n'], phi_fit['c50']*np.max(B_thetas), phi_r2))
    print('CV n = %.2f ; B50 = %.2f°, r^2 = %.2f' %
          (fit_cv['n'], fit_cv['c50']*np.max(B_thetas), cv_r2))

    make_NKR_monoaxis(figsize=figsize, do_save=True, filename=cluster_path,
                      cv_fit=fit_cv, cv_raw=CVs,
                      hwhh_fit=phi_fit, hwhh_raw=bthetas_fit,
                      B_thetas=B_thetas)
```
