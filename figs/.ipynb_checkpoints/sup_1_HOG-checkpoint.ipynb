# Supplementary Figure 1 : Histogram of oriented gradients 

```python
def generate_envelope(theta, b_theta, phase,
                      N_X, N_Y):

    fx, fy, ft = mc.get_grids(N_X, N_Y, 1)
    disk = mc.frequency_radius(fx, fy, ft) < .5

    mc_i = mc.envelope_gabor(fx, fy, ft,
                             V_X=0., V_Y=0., B_V=0.,
                             sf_0=.06, B_sf=.06,
                             theta=theta, B_theta=b_theta)
      

    return mc_i
```

```python
def generate_cloud(theta, b_theta, phase,
                   N_X, N_Y, seed, contrast=1.,
                   transition=False # obsolete param
                  ):


    mc_i = generate_envelope(theta, b_theta, phase,
                      N_X, N_Y)
      
    im_ = np.zeros((N_X, N_Y, 1))
    im_ += 2 * mc.rectif(mc.random_cloud(mc_i, seed=seed),
                     contrast=contrast)
    im_ += -.5
    return im_[:,:,0]
```

https://scikit-image.org/docs/dev/api/skimage.feature.html#skimage.feature.hog

```python
def norm_data(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
N_thetas = 60
N_thetas = 33

def compute_hog(image, N_thetas=N_thetas,feature_vector=False) :
    image = norm_data(image)
    fd = hog(image, orientations=N_thetas, pixels_per_cell=(64, 64),
                        cells_per_block=(1, 1), visualize=False, multichannel=True,
                       feature_vector = feature_vector, block_norm='L2', transform_sqrt = False)
    if feature_vector:
        return fd
    else:
        means = []
        for x in range(0, fd.shape[0]) :
            for y in range(0, fd.shape[1]) :
                means.append(fd[x,y,:,:,:])

        means = np.asarray(means)
        means = np.mean(means, axis = 0).ravel()
        return means
```

```python
def plot_hog(hog_vals, ax,
             roll, color, do_label, do_artifact) :
    hog_vals = np.roll(hog_vals, roll)
    #hog_vals = norm_data(hog_vals)
        
    xs = np.linspace(0, np.pi, len(hog_vals))
    ax.bar(xs, hog_vals, align = 'center', width = xs[1] - xs[0],
          facecolor = color, edgecolor = None, alpha = 1)
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    ax.grid(True, which = 'major', axis = 'y', linestyle = '--')
    ax.set_axisbelow(True)
    
    if do_label :
        ax.set_xticks([0, np.pi/4,  np.pi/2, 3*np.pi/4, np.pi])
        ax.set_xticklabels(['-90째', '-45째', r'$\theta_{0}$', '+45째', '+90째'])
        ax.tick_params(axis='both', labelsize=14)
        ax.set_xlabel('Orientation', fontsize = 14)
        ax.set_ylabel('Orientation distribution', fontsize = 14)
        
    else : 
        ax.set_xticks([0, np.pi/4,  np.pi/2, 3*np.pi/4, np.pi])
        ax.set_xticklabels([])
        #ax.set_yticklabels([])
        
    ax.set_ylim(np.min(hog_vals),
               np.max(hog_vals)+.01)
```
