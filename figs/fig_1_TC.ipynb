# Is the peak to baseline ratio different ?

```python
from scipy.stats import wilcoxon #nonparametric PAIRED version of the t-test. Continuity correctio applied
```

```python
changed = 0
ratios = []
for cluster_path in cluster_list :
    load_folder = '%s/%s' % (grouping_path, cluster_path)
    mean_FR = np.load(load_folder + '/0.000_plot_MC_TC_nonmerged_all.npy')
    pref_ori = []
    '''for ibt, bt in enumerate(B_thetas) :
        pref_ori.append(np.argmax(mean_FR[ibt]))
    if np.abs(pref_ori[0] - pref_ori[-1]) > 2 :''' 
    w, p = wilcoxon(np.max(mean_FR[-1,:,:], axis = 0)/np.min(mean_FR[-1,:,:], axis = 0),
                    np.max(mean_FR[0,:,:], axis = 0)/np.max(mean_FR[0,:,:], axis = 0),
                   alternative = 'less', zero_method = 'pratt', correction = True)
    ratios.append(np.mean(np.max(mean_FR[0,:,:], axis = 0) / np.min(mean_FR[0,:,:], axis = 0)) / np.mean(np.max(mean_FR[-1,:,:], axis = 0) / np.min(mean_FR[-1,:,:], axis = 0)))
    if p <.05 : changed +=1
    if cluster_path in important_guys : print(cluster_path, w, p)
        
print("Neurons that signif. changed amp from Bt0 to Bt36 :  %s / %s (%.2f percent)" % 
      (changed, len(cluster_list), changed/len(cluster_list)*100))
```

# How many are still tuned at BT = 36 ?

```python
changed = 0
list_changed = []
for cluster_path in cluster_list :
    load_folder = '%s/%s' % (grouping_path, cluster_path)
    mean_FR = np.load(load_folder + '/0.000_plot_MC_TC_nonmerged_all.npy')
    
    pref_ori = np.argmax(np.mean(mean_FR[-1,:,:], axis = -1))
    unpref_ori = pref_ori + 5
    if unpref_ori > 11 :
        unpref_ori = 0
        
    w, p = wilcoxon(mean_FR[0,pref_ori,:] , mean_FR[0,unpref_ori,:],
                   alternative = 'two-sided', zero_method = 'pratt', correction = True)
    if p <.05 : 
        changed +=1
        list_changed.append(cluster_path)
    if cluster_path in important_guys : print(cluster_path, w, p)
print("Neurons that are still tuned at Bt36 :  %s / %s (%.2f percent)" % 
      (changed, len(cluster_list), changed/len(cluster_list)*100))
```

# And which raw argmax is it for the bt0 and the last tuned (defined as the last to have a signif baseline/peak ratio)

```python
changed = 0
for cluster_path in cluster_list :
    load_folder = '%s/%s' % (grouping_path, cluster_path)
    mean_FR = np.load(load_folder + '/0.000_plot_MC_TC_nonmerged_all.npy')
    
    pref_ori = np.argmax(np.mean(mean_FR[-1,:,:], axis = -1))
    unpref_ori = pref_ori + 6
    if unpref_ori > 11 :
        unpref_ori = 0
        
    # Get the pvals of the diff between pref and unpref ori at all bthetas
    ps = []
    for ibt, bt in enumerate(B_thetas) :
        w, p = wilcoxon( mean_FR[ibt,pref_ori,:], 
                         mean_FR[ibt,unpref_ori,:])
        ps.append(p)
    # And the "last tuned" is the first <.05 = False in the list (reverse due to Bt array order) 
    try :
        last_tuned_idx = np.where(np.asarray(ps) < 0.05)[0][-1]
    except IndexError : #tuned all the way 
        last_tuned_idx = 0
        
    try :
        w, p = wilcoxon(np.argmax(mean_FR[-1,:,:], axis = 0),
                        np.argmax(mean_FR[last_tuned_idx,:,:], axis = 0),
                       alternative = 'two-sided', zero_method = 'pratt', correction = True)
        if p <.05 : changed +=1
    except ValueError :
        pass # wilcoxon returns a ValueError if x,y are exactly the same, in which case there's been no change
    if cluster_path in important_guys : print(cluster_path, w, p)
        
print("Neurons that signif. changed pref ori from Bt0 to their last tuned Bt:  %s / %s (%.2f percent)" % 
      (changed, len(cluster_list), changed/len(cluster_list)*100))
```

# TC on multiples axis

```python
def make_tc_multiaxes(figsize, do_save, filename,
                     fitted_TC, mean_FR,
                     thetas, B_thetas,
                     baseline,
                     mult_TC, mult_baseline,
                     do_legend, do_btlabs):
    
    fig, axs = plt.subplots(nrows = 8, figsize=figsize)

    fitted_TC *= mult_TC
    mean_FR *= mult_TC
    baseline *= mult_baseline

    for i, btheta in enumerate(B_thetas):
        ax = axs[i]
        xs = np.linspace(0, 180, len(fitted_TC[i]))

        ax.plot(xs, fitted_TC[i], c=colors[i], zorder=i,
                label=r'$B_\theta$ = %.2f°' % (B_thetas[i] * 180/np.pi))
        ax.fill_between(xs, np.min(mean_FR[i]), fitted_TC[i], color = colors[i],
                       zorder = 0, alpha = .6)
        ax.scatter(np.linspace(0, 180, 12), mean_FR[i], facecolor=colors[i],
                   zorder=i-.5, edgecolor=None, s=10)
        
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        
        ax.set_ylim(-2, np.max(fitted_TC[0]) + 0.2*np.max(fitted_TC[0]))
        
        

        #ax.axhline(baseline, linestyle='--', color='gray', zorder=-800)
    
        if i == 7 : 
            ax.set_xticks([0, 45, 90, 135, 180])
            ax.tick_params(axis='both', which='major', labelsize=13)
            ax.set_xlabel(r'$\theta$(°)', fontsize=16)
            ax.set_yticks([])
            ax.set_yticklabels([])
            
        elif i == 0 :
            y0 = (1/3) * np.max(fitted_TC[0]) 
            y1 =  (3/3) * np.max(fitted_TC[0])
            
            ax.plot([-5,-5], [y0, y1],
                   c = 'k', linewidth = 3)
            ax.text(x = -35, y = 1.5/3 * np.max(fitted_TC[0]), s = '%d sp.s$^{-1}$' % (2/3*np.max(fitted_TC[0])),
                   rotation = 90, fontsize = 12)
            
            ax.set_yticks([])
            ax.set_yticklabels([])
            ax.set_xticks([])
            ax.set_xticklabels([])
            
        else :
            ax.set_yticks([])
            ax.set_yticklabels([])
            
            ax.set_xticks([])
            ax.set_xticklabels([])
            
        if do_btlabs :
            ax.text(x = -110, y = .5 * np.max(fitted_TC[0]), s = r'$B_\theta$=%.1f°'% (B_thetas[::-1][i] * 180/np.pi),
                 color = colors[i], fontsize = 14, va = 'center', ha = 'center') 

        

    fig.subplots_adjust(hspace = 0)
    if do_legend:
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles[::-1], labels[::-1], frameon = False, fontsize = 14)
    if do_save:
        fig.savefig('./output/fig_1_TC_%s.pdf' %
                    filename, bbox_inches='tight', dpi=200, transparent=True)
    
    
    plt.show()
```
